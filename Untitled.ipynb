```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# (a) Plot of φ₂ vs σᴹ
plt.subplot(1, 2, 1)
phi2 = np.array([-0.189, -0.170, -0.135, 0, 0.153])  # V
sigma_M = np.array([-23.2, -16.0, -8.0, 0, 11.5])    # μC/cm²

plt.plot(phi2, sigma_M, 'bo-', label='0.01 M NaF')
plt.xlabel('φ₂ (V)')
plt.ylabel('σᴹ (μC/cm²)')
plt.title('(a) Plot of φ₂ vs σᴹ')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.legend()

# (b) Plot of φ₂ vs (E-Ez)
plt.subplot(1, 2, 2)
E_Ez = np.array([-1.4, -1.0, -0.5, 0, 0.5])  # V
phi2_b = np.array([-0.189, -0.170, -0.135, 0, 0.153])  # V

plt.plot(E_Ez, phi2_b, 'ro-', label='0.01 M NaF')
plt.xlabel('E-Ez (V)')
plt.ylabel('φ₂ (V)')
plt.title('(b) Plot of E-Ez vs φ₂')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.legend()

# Adjust layout
plt.tight_layout()

# Calculate linear regression parameters
slope_a, intercept_a, r_value_a, p_value_a, std_err_a = stats.linregress(phi2, sigma_M)
slope_b, intercept_b, r_value_b, p_value_b, std_err_b = stats.linregress(E_Ez, phi2_b)

# Print analysis results
print("Linear Regression Analysis:")
print("\n(a) φ₂ vs σᴹ:")
print(f"Slope: {slope_a:.2f} μC/cm²/V")
print(f"R² value: {r_value_a**2:.4f}")
print(f"y-intercept: {intercept_a:.4f}")

print("\n(b) E-Ez vs φ₂:")
print(f"Slope: {slope_b:.2f}")
print(f"R² value: {r_value_b**2:.4f}")
print(f"y-intercept: {intercept_b:.4f}")

# Show plot
plt.show()
```
